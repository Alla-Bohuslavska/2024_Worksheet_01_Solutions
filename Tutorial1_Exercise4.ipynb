```r
data(InsectSprays)
x <- InsectSprays$count[InsectSprays$spray=='A']
y <- InsectSprays$count[InsectSprays$spray=='C']
data <- data.frame(x,y)
colnames(data) <- c("Insecticide A", "Insecticide C")
boxplot(data, main = "Insect counts", ylab = "Number", col = c("lightskyblue3", "salmon"))
mean(x)
mean(y)
sd(x)
sd(y)

hist(x, freq=FALSE, col = "lightskyblue3", xlab="Insect count", ylab = "Relative frequency",
     main = "Insecticide A", xlim = c(0,25))
hist(y, freq=FALSE, col = "lightskyblue3", xlab="Insect count", ylab = "Relative frequency",
     main = "Insecticide C", xlim = c(0, 25))
```
